```python
%pylab inline
import os
userdoc = os.path.join(os.getcwd(),'Plots')
```

```python
likhoods = np.loadtxt('checksnaps_easier_likelihood.txt').reshape((5,5,5,5))
print likhoods.shape
a = np.array([0.5,0.75,1.0,1.25,1.5])
z0 = np.array([100,150,200,250,300])*1.0
fr = np.array([0.1,0.3,0.5,0.7,0.9])
rsn = np.array([0.8,0.9,1.0,2.0,3.0])*1.0e-3*15.0
```

```python
def confintervals(arr):
    logarr = np.log(arr)
    cont68 = np.exp(np.max(logarr)-0.5)
    cont95 = np.exp(np.max(logarr)-1.92)
    cont99 = np.exp(np.max(logarr)-5)
    return (cont68,cont95,cont99)

def fracscale():
    delta_r = (rsn[1]-rsn[0])/(rsn[-1]-rsn[0])
    delta_a = (a[1]-a[0])/(a[-1]-a[0])
    arr = np.zeros((z0.size,fr.size))
    for i in range(z0.size):
        for j in range(fr.size):
            arr[i,j]=delta_r*delta_a*np.sum(likhoods[:,i,j,:])
    
    return arr

def scalerate():
    delta_f = (fr[1]-fr[0])/(fr[-1]-fr[0])
    delta_a = (a[1]-a[0])/(a[-1]-a[0])
    arr = np.zeros((z0.size,rsn.size))
    for i in range(z0.size):
        for j in range(rsn.size):
            arr[i,j]= delta_f*delta_a*np.sum(likhoods[:,i,:,j])
            
    return arr

def scalealpha():
    delta_f = (fr[1]-fr[0])/(fr[-1]-fr[0])
    delta_r = (rsn[1]-rsn[0])/(rsn[-1]-rsn[0])
    arr = np.zeros((a.size,z0.size))
    for i in range(a.size):
        for j in range(z0.size):
            arr[i,j]= delta_f*delta_r*np.sum(likhoods[i,j,:,:])
    
    return arr

def alphafrac():
    delta_z = (z0[1]-z0[0])/(z0[-1]-z0[0])
    delta_r = (rsn[1]-rsn[0])/(rsn[-1]-rsn[0])
    arr = np.zeros((a.size,fr.size))
    for i in range(a.size):
        for j in range(fr.size):
            arr[i,j]= delta_r*delta_z*np.sum(likhoods[i,:,j,:])
            
    return arr

def fracrate():
    delta_a = (a[1]-a[0])/(a[-1]-a[0])
    delta_z = (z0[1]-z0[0])/(z0[-1]-z0[0])
    arr = np.zeros((fr.size,rsn.size))
    for i in range(fr.size):
        for j in range(rsn.size):
            arr[i,j]= delta_a*delta_z*np.sum(likhoods[:,:,i,j])
            
    return arr

def ratealpha():
    delta_z = (z0[1]-z0[0])/(z0[-1]-z0[0])
    delta_f = (fr[1]-fr[0])/(fr[-1]-fr[0])
    arr = np.zeros((a.size,rsn.size))
    for i in range(a.size):
        for j in range(rsn.size):
            arr[i,j]= delta_z*delta_f*np.sum(likhoods[i,:,:,j])
            
    return arr
```

__Scale Height vs Fraction__

```python
import contourplots as cplt
reload(cplt)
lik = fracscale()

fnew,tnew,lnew = cplt.interpvals(fr,z0,lik)
lnew_68,lnew_95,lnew_99 = confintervals(lnew)
plt.figure(figsize=(8,6))
plt.contourf(fnew,tnew,lnew,cmap='autumn_r')
plt.xlabel('Fraction',fontsize=20)
plt.ylabel('Scale Height [pc]',fontsize=20)
ax = plt.gca()
ax.tick_params(labelsize=15)
plt.colorbar(format='%.1e')
#plt.savefig(os.path.join(userdoc,'ScaleFraction.png'))
plt.show()
```

__Scale Height vs SN Rate__

```python
lik_z0rsn = scalerate()

rnew,znew,lnew = cplt.interpvals(rsn,z0,lik_z0rsn)
plt.figure(figsize=(8,6))
plt.contourf(rnew/(15.0*1.0e-3),znew,lnew,cmap='autumn_r')
plt.xlabel(r'SN Rate $[\times 10^{-3} yr^{-1}]$',fontsize=20)
plt.ylabel('Scale Height [pc]',fontsize=20)
#plt.xlim(0.8,1.5)
ax = plt.gca()
ax.tick_params(labelsize=15)
plt.colorbar(format='%.1e')
plt.savefig(os.path.join(userdoc,'ScaleRate.png'))
plt.show()
```

__Scale Height vs $\alpha$__

```python
lik_z0a = scalealpha()

znew,anew,lnew = cplt.interpvals(z0,a,lik_z0a)
plt.figure(figsize=(8,6))
plt.contourf(znew,anew,lnew,cmap='autumn_r')
plt.ylabel(r'$\alpha$',fontsize=20)
plt.xlabel('Scale Height [pc]',fontsize=20)
ax = plt.gca()
ax.tick_params(labelsize=15)
plt.colorbar(format='%.1e')
plt.savefig(os.path.join(userdoc,'AlphaScale.png'))
plt.show()
```

__$\alpha$ vs Fraction__

```python
lik_af = alphafrac()

fnew,anew,lnew = cplt.interpvals(fr,a,lik_af)
plt.figure(figsize=(8,6))
plt.contourf(fnew,anew,lnew,cmap='autumn_r')
plt.ylabel(r'$\alpha$',fontsize=20)
plt.xlabel('Fraction',fontsize=20)
ax = plt.gca()
ax.tick_params(labelsize=15)
plt.colorbar(format='%.1e')
plt.savefig(os.path.join(userdoc,'AlphaFraction.png'))
plt.show()
```

```python
lik_fr = fracrate()

rnew,fnew,lnew = cplt.interpvals(rsn,fr,lik_fr)
plt.figure(figsize=(8,6))
plt.contourf(rnew/(15.0*1.0e-3),fnew,lnew,cmap='autumn_r')
plt.xlabel(r'SN Rate $[\times 10^{-3} yr^{-1}]$',fontsize=20)
plt.ylabel('Fraction',fontsize=20)
#plt.xlim(0.8,1.5)
ax = plt.gca()
ax.tick_params(labelsize=15)
plt.colorbar(format='%.1e')
plt.savefig(os.path.join(userdoc,'FractionRate.png'))
plt.show()
```
