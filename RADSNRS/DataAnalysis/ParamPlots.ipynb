```python
likhoods = np.loadtxt('epsb_likhood_lum.txt').reshape((10,10,10))
#targetfile = 'TrianglePlot_diams2.png'
print likhoods.shape
z0 = np.linspace(10,900,10)*1.0
rsn = np.linspace(1.0,5.0,10)*1.0e-3*15.0
epsb = np.logspace(-3,-1.3,10)
delta_r = (rsn[1]-rsn[0])/(rsn[-1]-rsn[0])
delta_z = (z0[1]-z0[0])/(z0[-1]-z0[0])
delta_b = (epsb[1]-epsb[0])/(epsb[-1]-epsb[0])
norm = np.sum(likhoods)*delta_r*delta_z*delta_b
print norm
shades = ('w',(0.0,0.8,1.0),'b')
```

```python
def confintervals(arr):
    logarr = np.log(arr)
    cont68 = np.exp(np.max(logarr)-0.5)
    cont95 = np.exp(np.max(logarr)-1.92)
    cont99 = np.exp(np.max(logarr)-5)
    return (cont68,cont95,cont99)

def scalerate():
    arr = np.zeros((z0.size,rsn.size))
    for i in range(z0.size):
        for j in range(rsn.size):
            arr[i,j]= delta_b*np.sum(likhoods[i,j,:])
            
    return arr/norm

def scaleepsb():
    arr = np.zeros((z0.size,epsb.size))
    for i in range(z0.size):
        for j in range(epsb.size):
            arr[i,j]= delta_r*np.sum(likhoods[i,:,j])
            
    return arr/norm

def rateepsb():
    arr = np.zeros((rsn.size,epsb.size))
    for i in range(rsn.size):
        for j in range(epsb.size):
            arr[i,j]=delta_z*np.sum(likhoods[:,i,j])
            
    return arr/norm
```
