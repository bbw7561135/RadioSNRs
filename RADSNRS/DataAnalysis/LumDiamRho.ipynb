```python
lum = np.loadtxt('paramtst_lumhist.txt')
diam = np.loadtxt('paramtst_diamhist.txt')
dens = np.loadtxt('paramtst_denshist.txt')
profs = np.loadtxt('paramtst_profhist.txt')
print lum.size, diam.size, dens.size
print np.count_nonzero(lum)
print np.count_nonzero(diam)
print np.count_nonzero(dens)
print np.count_nonzero(profs)
```

```python
path = '../Inputs/'
lmclums = (1.0e24)*1.2*1000.0*np.loadtxt(path+'lmcradiolumscopy.txt')*0.05*0.05
lmcdiams_temp = np.loadtxt(path+'lmc_angdiams.txt')
print lmcdiams_temp
lmcdiams = ((lmcdiams_temp*np.pi)/(180.0*3600.0))*0.05*(1.0e6)
lmcdens = np.loadtxt(path+'lmc_snrs_opaccopy.txt')
```

```python
lum = lum[np.nonzero(lum)]
diam = diam[np.nonzero(diam)]
dens = dens[np.nonzero(dens)]
profs = profs[np.nonzero(profs)]

fig7 = plt.figure(7)
ax7 = fig7.add_subplot(111)
fig7.set_size_inches(7,5)

lum_1a = lum[np.where(profs==7.0)[0]]
lum_cc = lum[np.where(profs==12.0)[0]]
rho_1a = dens[np.where(profs==7.0)[0]]
rho_cc = dens[np.where(profs==12.0)[0]]
ax7.plot(lmcdens,lmclums,'ko',label='Observed')
ax7.plot(rho_1a,lum_1a,'bo',mew=1.5,mec='b',mfc='none',label='Model-Ia')
ax7.plot(rho_cc,lum_cc,'ro',mew=1.5,mec='r',mfc='none',label='Model-CC')
ax7.set_xscale('log')
ax7.set_yscale('log')
ax7.set_xlabel(r'$<\rho> [cm^{-2}]$',fontsize='20')
ax7.set_ylabel(r'Radio Luminosity [ergs/s/Hz]',fontsize='20')
for axis in ['top','bottom','left','right']:
  ax7.spines[axis].set_linewidth(2.0)
plt.tick_params(axis='both',length=13,width=2,labelsize='18',which='major')
plt.tick_params(axis='both',length=6,width=1,labelsize='18',which='minor')
#ax7.axis([1.0e18,1.0e23,1.0e19,1.0e26])                                                                                                                      
ax7.legend(loc=2,numpoints=1,fontsize='15',borderaxespad=1.)
plt.tight_layout()
plt.savefig('lumrho.png')

```

```python
fig8 = plt.figure(8)
fig8.set_size_inches(7,5)
ax8 = fig8.add_subplot(111)
rad_1a = diam[np.where(profs==7.0)[0]]
rad_cc = diam[np.where(profs==12.0)[0]]
rho_1a = dens[np.where(profs==7.0)[0]]
rho_cc = dens[np.where(profs==12.0)[0]]
ax8.plot(lmcdens,lmcdiams,'ko',label='Observed')
ax8.plot(rho_1a,rad_1a,'bo',mew=1.5,mec='b',mfc='none',label='Model-Ia')
ax8.plot(rho_cc,rad_cc,'ro',mew=1.5,mec='r',mfc='none',label='Model-CC')
ax8.set_xscale('log')
ax8.set_yscale('log')
ax8.set_xlabel(r'$<\rho_{col}> [cm^{-2}]$',fontsize='20')
ax8.set_ylabel(r'SNR Diameter [pc]',fontsize='20')
#ax8.axis([1.0e18,1.0e23,1.0,200.0])                                                                                                                          
ax8.legend(loc=3,numpoints=1,fontsize='15',borderaxespad=1.)
ax8.set_ylim(0.3,1000)
for axis in ['top','bottom','left','right']:
  ax8.spines[axis].set_linewidth(2.0)
plt.tick_params(axis='both',length=13,width=2,labelsize='18',which='major')
plt.tick_params(axis='both',length=6,width=1,labelsize='18',which='minor')
plt.tight_layout()
plt.savefig('diamrho.png')
```
