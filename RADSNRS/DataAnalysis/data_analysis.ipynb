Create Cumulative (or differential) histogram of the luminosities (observed vs. model) for each $(f,d)$

The __Maximum Likelihood Method__ as described in (Badenes 2010). We divide our sample into equal-sized logarithmic bins. The probability of finding $j$ remnants in the bin, given the expectation $\langle n \rangle$ from the model with parameters $(f,d)$ is a Poisson probability defined as:-

$p(j|n) = \dfrac{\mathrm{e}^{-n} n^j}{j!}$

From this we can calculate a likelihood :-

$\mathrm{ln}\ L = \sum_{i=1}^{N_{bins}} \mathrm{ln}\ p_i$

And hopefully this can be used to construct confidence contours

```python
import math as mt

#CALCULATE N(OBS) AND N(MODEL) PER BIN

l_cutoff = 2.0e23
obs_bins = np.logspace(np.log10(l_cutoff),np.log10(lmclums_temp.max()),6)
bincen = (obs_bins[1:]+obs_bins[:-1])
print obs_bins
plt.hist(lmclums_temp,bins=obs_bins,histtype='step',lw=2,color='g',label='Observed')
n,bins = np.histogram(lmclums_temp,bins=obs_bins)
#patches[0].set_xy(patches[0].get_xy()[1:])
n2 = np.zeros((lum.shape[0],obs_bins.size-1))
for ind,lums in enumerate(lum):
    lums = lums[np.nonzero(lums)]
    #plt.hist(lums,bins=obs_bins,histtype='step',color='k',alpha=0.5)
    n2[ind], bins2 = np.histogram(lums,bins=obs_bins)
    
#For the other set of lums 
n3 = np.zeros((lum2.shape[0],obs_bins.size-1))
for ind,lums in enumerate(lum2):
    lums = lums[np.nonzero(lums)]
    #plt.hist(lums,bins=obs_bins,histtype='step',color='k',alpha=0.5)
    n3[ind], bins3 = np.histogram(lums,bins=obs_bins)

#avg_n = np.array([np.mean(col) for col in np.transpose(n2)])
#std_n = np.array([np.std(col) for col in np.transpose(n2)])

avg_n = np.mean(n2,axis=0)
std_n = np.std(n2,axis=0)
avg_n2 = np.mean(n3,axis=0)
std_n2 = np.std(n3,axis=0)
#plt.errorbar(bincen,avg_n,yerr=std_n,fmt='o',label='Model')
plt.errorbar(bincen*0.5,avg_n/10.0,yerr=std_n/10.0,fmt='o',ms=7,color='r',label=r'$10R_{SN}$')
plt.errorbar(bincen*0.45,avg_n2,yerr=std_n2,fmt='o',ms=7,color='b',label=r'$R_{SN}$')

#CALCULATING LIKELIHOOD
def poissonProb(j,n):
    return (np.exp(-n)*(n**j))/mt.factorial(j)

likhood_temp = np.array([poissonProb(n[ind],avg_n[ind]) for ind in range(n.size)])
likhood = np.prod(likhood_temp)

print likhood
plt.xscale('log')
plt.xlabel('Radio Luminosity [ergs/s/Hz]',fontsize=15)
plt.ylim(0,20)
fig = plt.gcf()
fig.set_size_inches(8,7)
plt.tick_params(axis='both',which='major',labelsize=15)
plt.xlim(8.0e22,3.0e25)
plt.legend()
plt.show()
print np.vstack([n,avg_n,std_n])
```
