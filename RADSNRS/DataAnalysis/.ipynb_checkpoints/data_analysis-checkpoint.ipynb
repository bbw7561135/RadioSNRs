Create Cumulative (or differential) histogram of the luminosities (observed vs. model) for each $(f,d)$

The __Maximum Likelihood Method__ as described in (Badenes 2010). We divide our sample into equal-sized logarithmic bins. The probability of finding $j$ remnants in the bin, given the expectation $\langle n \rangle$ from the model with parameters $(f,d)$ is a Poisson probability defined as:-

$p(j|n) = \dfrac{\mathrm{e}^{-n} n^j}{j!}$

From this we can calculate a likelihood :-

$\mathrm{ln}\ L = \sum_{i=1}^{N_{bins}} \mathrm{ln}\ p_i$

And hopefully this can be used to construct confidence contours

```python
import math as mt

l_cutoff = 0.1 #ergs/s/Hz        

obs_bins = np.linspace(0,lmcdiams_temp.max(),6)
n,bins,patches =plt.hist(lmcdiams_temp,bins=obs_bins,histtype='step',lw=5,color='g',label='Observed')
n2 = np.zeros((lum.shape[0],obs_bins.size-1))
for ind,lums in enumerate(lum):
    lums = lums[np.nonzero(lums)]
    plt.hist(lums,bins=obs_bins,histtype='step',color='k',alpha=0.5)
    n2[ind], bins2 = np.histogram(lums,bins=obs_bins)

#plt.xscale('log')
plt.xlabel('Diameter [pc]',fontsize=15)
plt.ylim(0,200)
fig = plt.gcf()
fig.set_size_inches(8,7)
plt.legend()
plt.show()

avg_n = np.array([np.mean(col) for col in np.transpose(n2)])
std_n = np.array([np.std(col) for col in np.transpose(n2)])
print np.vstack([n,avg_n])

#CALCULATING LIKELIHOOD
def poissonProb(j,n):
    return (np.exp(-n)*(n**j))/mt.factorial(j)

likhood_temp = np.array([poissonProb(n[ind],avg_n[ind]) for ind in range(n.size)])
print likhood_temp
likhood = np.prod(likhood_temp)

print likhood
```

```python
import math as mt

#CALCULATE N(OBS) AND N(MODEL) PER BIN

l_cutoff = 2.0e23
obs_bins = np.logspace(np.log10(l_cutoff),np.log10(lmclums_temp.max()),6)
n,bins,patches =plt.hist(lmclums_temp,bins=obs_bins,histtype='step',lw=5,color='g',label='Observed')
#patches[0].set_xy(patches[0].get_xy()[1:])
n2 = np.zeros((lum.shape[0],obs_bins.size-1))
for ind,lums in enumerate(lum):
    lums = lums[np.nonzero(lums)]
    plt.hist(lums,bins=obs_bins,histtype='step',color='k',alpha=0.5)
    n2[ind], bins2 = np.histogram(lums,bins=obs_bins)

avg_n = np.array([np.mean(col) for col in np.transpose(n2)])
std_n = np.array([np.std(col) for col in np.transpose(n2)])

#CALCULATING LIKELIHOOD
def poissonProb(j,n):
    return (np.exp(-n)*(n**j))/mt.factorial(j)

likhood_temp = np.array([poissonProb(n[ind],avg_n[ind]) for ind in range(n.size)])
print likhood_temp
likhood = np.prod(likhood_temp)

print likhood
plt.xscale('log')
plt.xlabel('Radio Luminosity [ergs/s/Hz]',fontsize=15)
plt.ylim(0,30)
fig = plt.gcf()
fig.set_size_inches(8,7)
plt.legend()
plt.show()
print np.vstack((n,avg_n))
```

## Parameter Space (d-$f$)

```python
from scipy.interpolate import interp2d

thick = np.loadtxt('checksnaps_easier_thickness.txt')
snrate = np.loadtxt('checksnaps_easier_snrate.txt')
likelihood = np.loadtxt('checksnaps_easier_likelihood.txt').reshape(thick.size,frac.size)

S,T = np.meshgrid(snrate,thick)
np.set_printoptions(precision=4)
print S,'\n',T,'\n'
thicknew = np.linspace(thick.min(),thick.max(),500)
snratenew = np.linspace(snrate.min(),snrate.max(),500)
f = interp2d(snrate,thick,likelihood,kind='linear')
likhood_new = f(snratenew,thicknew)
print likelihood
```

```python
import os
plt.figure(figsize=(7,6))
plt.contourf(snratenew*1000,thicknew,likhood_new,cmap='autumn_r')
#plt.title('5x5 (f,d), ks-test,cutoff=2.0e23',fontsize=18)
plt.xlabel('SN Rate (Ia/CC)',fontsize=20)
plt.ylabel('Scale Height[pc]',fontsize=20)
ax = plt.gca()
ax.tick_params(labelsize=15)
plt.colorbar(format='%.1e')
#userdoc = os.path.join(os.getcwd(),'Plots')
#plt.savefig(os.path.join(userdoc,'7'))
#plt.ylim(1,30)
plt.show()
print likelihood
```
