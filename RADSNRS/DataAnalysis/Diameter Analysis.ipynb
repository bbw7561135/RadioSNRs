## Differential Histogram

```python
import math as mt

l_cutoff = 0.1 #ergs/s/Hz        

obs_bins = np.linspace(0,lmcdiams_temp.max(),6)
n,bins,patches =plt.hist(lmcdiams_temp,bins=obs_bins,histtype='step',lw=5,color='g',label='Observed')
n2 = np.zeros((lum.shape[0],obs_bins.size-1))
for ind,lums in enumerate(lum):
    lums = lums[np.nonzero(lums)]
    plt.hist(lums,bins=obs_bins,histtype='step',color='k',alpha=0.5)
    n2[ind], bins2 = np.histogram(lums,bins=obs_bins)

#plt.xscale('log')
plt.xlabel('Diameter [pc]',fontsize=15)
plt.ylim(0,200)
fig = plt.gcf()
#fig.set_size_inches(8,7)
plt.legend()
plt.show()

avg_n = np.array([np.mean(col) for col in np.transpose(n2)])
std_n = np.array([np.std(col) for col in np.transpose(n2)])
print np.vstack([n,avg_n])

#CALCULATING LIKELIHOOD
def poissonProb(j,n):
    return (np.exp(-n)*(n**j))/mt.factorial(j)

likhood_temp = np.array([poissonProb(n[ind],avg_n[ind]) for ind in range(n.size)])
print likhood_temp
likhood = np.prod(likhood_temp)

print likhood
```

```python
from scipy.interpolate import interp2d

thick = np.loadtxt('checksnaps_easier_thickness.txt')
frac = np.loadtxt('checksnaps_easier_fraction.txt')
likelihood = np.loadtxt('checksnaps_easier_likelihood.txt').reshape(thick.size,frac.size)

F,T = np.meshgrid(frac,thick)
np.set_printoptions(precision=4)
#print T,'\n',F,'\n'
thicknew = np.linspace(thick.min(),thick.max(),500)
fracnew = np.linspace(frac.min(),frac.max(),500)
f = interp2d(frac,thick,likelihood,kind='linear')
likhood_new = f(fracnew,thicknew)
print T[0:3]
print F[0:3]
print likelihood[0:3]
```
